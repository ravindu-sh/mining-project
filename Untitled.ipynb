```python
import asyncio
import aiohttp
from aiolimiter import AsyncLimiter
from bs4 import BeautifulSoup as bs

limiter = AsyncLimiter(5, 1)  # 5 requests per second (adjust as needed)

async def search_company(session, search_term):
    global limiter
    URL = "https://stockanalysis.com/api/search"
    params = {"q": str(search_term)}
    headers = {
        "accept": "*/*",
        "accept-encoding": "gzip, deflate, br, zstd",
        "accept-language": "en-US,en;q=0.9",
        "priority": "u=1, i",
        "referer": "https://stockanalysis.com/",
        "sec-ch-ua": '"Not:A-Brand";v="99", "Brave";v="145", "Chromium";v="145"',
        "sec-ch-ua-mobile": "?0",
        "sec-ch-ua-platform": '"Linux"',
        "sec-fetch-dest": "empty",
        "sec-fetch-mode": "cors",
        "sec-fetch-site": "same-origin",
        "sec-gpc": "1",
        "user-agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36",
    }
    async with limiter:
        async with session.get(URL, headers=headers, params=params) as response:
            results = await response.json()
    if not results.get('data'):
        return None
    item = results['data'][0]
    return f"https://stockanalysis.com/{'quote' if item['t'] == 'sy' else 'stocks'}/{item['s']}/company"

def get_website(soup):
    def has_noopener_nofollow(tag):
        if tag.name != "a":
            return False
        rel = tag.get("rel", [])
        if isinstance(rel, str):
            rel = rel.split()
        return "noopener" in rel and "nofollow" in rel
    anchors = soup.find_all(has_noopener_nofollow)
    if not anchors:
        return False
    url = anchors[0].get('href')
    return url

async def get_company_url(session, name):
    global limiter
    stock_website_for_company = await search_company(session, name)
    if not stock_website_for_company:
        return None
    async with limiter:
        async with session.get(stock_website_for_company) as website:
            text = await website.text()
    soup = bs(text, 'html.parser')
    url = get_website(soup)
    return url


async def process_row(session, idx, row, df):
    name = row.get('Name')
    url = row.get("URL")
    if not pd.isna(url):
        print(f"already exists {name} -> {url}")
        return
        # return None, None
    try:
        website = await get_company_url(session, name)
        print(f"found {name} -> {website}")
        df.at[idx, 'URL'] = website
        # # return idx, website
    except Exception as e:
        print(f"failed {name} {e}")
        # return None, None

async def main():
    async with aiohttp.ClientSession() as session:
        tasks = [
            process_row(session, idx, row, df)
            for idx, row in df.iterrows()
        ]
        results = await asyncio.gather(*tasks)
        # for idx, website in results:
        #     if website:
        #         df.at[idx, "URL"] = website

await main()
```
